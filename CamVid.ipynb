```python
import os
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tensorflow as tf
import re
```

```python
df = pd.read_csv('class_dict.csv')
```

```python
train_img = {'images' : [],
            'label_img' : []}
val_img =  {'images' : [],
            'label_img' : []}

def load_data(frameObj = None,img_path = None,label_path = None,size = (256,256)):
    img_names = os.listdir(img_path)
    label_names = []
    
    for name in img_names:
        label_names.append(re.sub('\.png', '_L.png', name))
        
    img_address = img_path + '/'
    label_address = label_path + '/'
    
    for i in range(len(img_names)):
        img = cv2.imread(img_address + img_names[i] )
        label = cv2.imread(label_address +label_names[i] )
        
        img = cv2.resize(img,size)
        label = cv2.resize(label,size)
        
        frameObj['images'].append(img)
        frameObj['label_img'].append(label)
        
    return frameObj
    
        
```

```python
train_data = load_data(train_img,img_path = 'train',label_path = 'train_labels',size=(256,256))
```

```python
val_data = load_data(val_img,img_path = 'val',label_path = 'val_labels',size=(256,256))
```
